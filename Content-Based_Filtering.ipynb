# Content-Based Filtering

```
import pandas as pd

# movies = pd.read_csv('movies_small.csv', sep=';')
movies = pd.read_csv('movies.csv')
```

```
movies
```

```
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english')

#  Convert null values to empty strings
movies['overview'] = movies['overview'].fillna('')
```

```
movies['overview']
```

```
tfidf_matrix = tfidf.fit_transform(movies['overview'])
```

```
pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf.get_feature_names())
```

```
tfidf_matrix.shape
```

### Similarity Matrix

```
from sklearn.metrics.pairwise import linear_kernel

similarity_matrix = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```
similarity_matrix
```

```
similarity_matrix.shape
```

### Find the most similar movies to a certain movie

```
movie_title = 'Cars 2'
```

```
idx = movies.loc[movies['title'] == movie_title].index[0]
idx
```

```
scores = list(enumerate(similarity_matrix[idx]))
scores
```

```
# scores = [(score, index) for index, score in scores]
# scores
```

```
# scores = sorted(scores, reverse=True)
scores = sorted(scores, key=lambda x: x[1], reverse=True)
# Remove first movie (itself)
scores
```

```
movies_indices = [tpl[0] for tpl in scores[1:4]]
movies_indices
```

```
list(movies['title'].iloc[movies_indices])
```

```
def similar_movies(movie_title, no_of_movies):
    idx = movies.loc[movies['title'] == movie_title].index[0]
    scores = list(enumerate(similarity_matrix[idx]))
    scores = sorted(scores, key=lambda x: x[1], reverse=True)
    movies_indices = [tpl[0] for tpl in scores[1:no_of_movies + 1]]
    movies_list = list(movies['title'].iloc[movies_indices])
    return movies_list
```

```
similar_movies('Kung Fu Panda 3', 3)
```
