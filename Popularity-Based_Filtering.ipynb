## Popularity-Based Filtering

### Load the data

```
import pandas as pd

movies = pd.read_csv('movies.csv')
credits = pd.read_csv('credits.csv')
ratings = pd.read_csv('ratings.csv')
```

### Calculate a weighted rating

WR = (v / (v + m)) * R + (m + (v + m)) * C

v - no. of votes (we have)

m - min. no. of votes to be popular (to calculate)

R - avg. rating (we have)

C - avg rating across all movies (to calculate)

```
m = movies['vote_count'].quantile(0.9)
m
```

```
c = movies['vote_average'].mean()
c
```

```
movies_filtered = movies.copy().loc[movies['vote_count'] >= m]
```

```
def weighted_rating(df, m=m, c=c):
    v = df['vote_count']
    R = df['vote_average']
    # Calculation based on the IMDB formula
    wr = ((v / (v + m)) * R) + ((m / (m + v)) * c)
    return wr
```

```
movies_filtered['weighted_rating'] = movies_filtered.apply(weighted_rating, axis=1)
```

```
movies_filtered
```

```
movies['weighted_rating'] = movies.apply(weighted_rating, axis=1)
```

```
movies
```

```
movies_filtered.sort_values('weighted_rating', ascending=False).head(10)
```
